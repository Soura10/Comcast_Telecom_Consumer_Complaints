## Task1:Provide the trend chart for the number of complaints at monthly and daily granularity levels.

```python
df["Month"]=pd.to_datetime(df["Date_month_year"]).dt.month_name()
```

```python
df["Date"]=pd.to_datetime(df["Date_month_year"]).dt.day
```

```python
#Graph for Month wise
df.groupby(["Month"])["Customer Complaint"].count().plot()
```

```python
#Graph for Date wise
df.groupby(["Date"])["Customer Complaint"].count().plot(kind="bar",color="green")
```

Create a new categorical variable with value as Open and Closed. Open & Pending is to be categorized as Open and Closed & Solved is to be categorized as Closed.

```python
df.Status.unique()
```

```python
#df["Consolidated Status"]=["Open" if Status=="Open" or Status=="Pending" else "Closed" for Status in df["Status"]]
df['Status']=df['Status'].apply(lambda x: 'Open' if ((x=='Open') | (x=='Pending')) else 'Closed')
```

```python
df.head()
```

Provide state wise status of complaints in a stacked bar chart. Use the categorized variable from Q3. Provide insights on

```python
df.groupby(["State"]).size().sort_values(ascending=False).to_frame().reset_index()
```

```python
df.groupby(["State"]).size().sort_values(ascending=False).to_frame().reset_index().rename({0:"Count"},axis=1)[:5]
```
